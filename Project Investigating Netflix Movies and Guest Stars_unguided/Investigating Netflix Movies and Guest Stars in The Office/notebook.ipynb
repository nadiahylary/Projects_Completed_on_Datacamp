```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

office_episodes_df = pd.read_csv('datasets/office_episodes.csv')

office_episodes_df.head()
```

# One way to visualize the popularity of the different episodes is:

```python
colors = []                                 
for lab, row in office_episodes_df.iterrows():
    if row['scaled_ratings']   < 0.25:
        colors.append("red")
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        colors.append("orange")
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        colors.append("lightgreen")
    elif row['scaled_ratings'] >= 0.75:
        colors.append("darkgreen")
                                    
#print(colors[:25])
#office_episodes_df[['ratings']]
#office_episodes_df[['scaled_ratings']]
```

```python
sizes = []

for lab, row in office_episodes_df.iterrows():
    if row['has_guests']:
        sizes.append(250)
    else:
        sizes.append(25)
#print(sizes[:10])
office_episodes_df['colors'] = colors
office_episodes_df['sizes'] = sizes
#office_episodes_df.head()
```

# Another way to visualize the popularity of the different episodes is:

```python
fig = plt.figure()

non_guest_df = office_episodes_df[office_episodes_df['has_guests'] == False]
guest_df = office_episodes_df[office_episodes_df['has_guests'] == True] 

plt.rcParams['figure.figsize'] = [12, 8]

episodes = office_episodes_df['episode_number']
views = office_episodes_df['viewership_mil']
plt.scatter(x = non_guest_df.episode_number, y = non_guest_df.viewership_mil, s = non_guest_df['sizes'], c = non_guest_df['colors'] ) # , s = sizes, c = colors, edgecolor = 'black', linewidth = 1
plt.scatter(x = guest_df.episode_number, y = guest_df.viewership_mil, s = guest_df['sizes'], c = guest_df['colors'] ) # , s = sizes, c = colors, edgecolor = 'black', linewidth = 1
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number", fontsize=18)
plt.ylabel("Viewership (Millions)", fontsize=18)

plt.show()
```

**Episode number:**

```python
# Highest number of viewership i.e the outlier:
most_views = np.amax(np.array(office_episodes_df['viewership_mil']))
# Episode number
office_episodes_df[office_episodes_df['viewership_mil'] == most_views].episode_number
```
